```python
import matplotlib.pylab as plt # I use version 3.1.2
import numpy as np # I use version 1.18.1
import cv2
import math

def fixColor(image):
    return(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python

# import math
def calculateDistance(x1,y1,x2,y2):
    dist = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return dist

count = 3

while (True):
    img_name = str(count) + '.jpg'
    count = count - 1
    
    img = cv2.imread(img_name)
    img = cv2.rotate(img, cv2.ROTATE_180)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    
    
    canny = cv2.Canny(blurred, 40, 140) #this is an OpenCV function that helps in detection of edges, the 1st argument is the 
    y_axis, x_axis = np.where(canny == 255)
    min = np.min(y_axis) #taking out the min value of y axis 
    max = np.max(y_axis) #taking out the max value of y axis

    
    if len(x_axis[np.where(y_axis == math.ceil(max/2))]) > 3: #forward movement logic starts here
        print('Moving forward because wall is too far in image: ', img_name)    
    else:
        print('Now performing wall detection on image ', img_name, 'because wall is near now')
        plt.imshow(fixColor(canny))
        break  #forward movement logic ends here

uy_axis = y_axis[0];
upperx_axis = x_axis[0];
final_x = 0
print(upperx_axis)

for i in range(max, min, -10):
    print(i)
    arr_size = len(x_axis[np.where(y_axis == i)])
    lowx_axis = x_axis[np.where(y_axis == i)]
    midx_value = x_axis[np.where(y_axis == (max/2))]

    for k in range(0 , arr_size):
        if midx_value[k] > upperx_axis - 100 and midx_value[k] <= upperx_axis + 100:
            if lowx_axis[k] > upperx_axis - 100 and lowx_axis[k] <= upperx_axis + 100:
                final_x = lowx_axis[k]
                final_y = i
                print(lowx_axis[k], 'This is the value')
        else:
            print('We have found a break in a line this is not the line you are looking for')
        

    if final_x > 0:
        break;
if final_x > np.max(x_axis)/2:
    print('Your Turning is on the right since the lenght of edge on right side is: ', calculateDistance(final_x,final_y, upperx_axis, uy_axis))
else:
    print('your turning is on the left')

        
        
        
if len(x_axis[np.where(y_axis == math.ceil(max/2))]) == 3: #math.ceil function round up the number make it whole value since there are not decimal numbers in y_axis or x_axis
    #once we round it up we need to find the coordinate where the middle value of y_axis lie in y-axis variable this is done using np.where
    #once we have the coordinates of where middle value lie we use these coordinate to fetch data from x axis variable and we check the lenght of x axis data we have retrived
    #this is done so that we can figure out our problem of detecting wall according to what a robot can see from it is standing 
    num_edge = 3
else:
    num_edge = 2

for i in range(min, max, 10): # run the loop from lowest to height value in y_axis variable with step size 10
    if len(x_axis[np.where(y_axis == i)]) == num_edge: #in the loop check the number of values received againts each y-axis value, if value is equal to either value set in num_edges then fetch out the coordinates
    
        y = i #fetch out y coordinate of the the required edge
        x = np.sort(x_axis[np.where(y_axis == i)]) #fetch out x coordinate of the the required edge
        break;   #if edge requirment is satisfied then break the loop so that we do not overwrite y and x variable
#print(x, y)
canny= canny[y-20:1000, x[0]-10:x[1]+10] #crop the image by defining the part we need to show we have subtracted 20 in y coordinate so that we can have correct visual of the edge and similarly subtracted from lower value in x coordinate and added in heigher value of x coordinate
plt.imshow(fixColor(canny))



```
